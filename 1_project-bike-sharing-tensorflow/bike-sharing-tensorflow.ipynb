# 共享单车使用量预测

利用优达学成提供的数据，对共享单车的使用量进行预测。本项目采用神经网络模型。


```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 加载、准备数据

用pandas载入数据，看一下数据的大致结构

```python
data_path = 'Bike-Sharing-Dataset/hour.csv'

rides = pd.read_csv(data_path)
```

## 检查数据

下图展示的是数据集中前 10 天左右的骑车人数（某些天不一定是 24 个条目，所以不是精确的 10 天）。你可以在这里看到每小时租金。这些数据很复杂！周末的骑行人数少些，工作日上下班期间是骑行高峰期。我们还可以从上方的数据中看到温度、湿度和风速信息，所有这些信息都会影响骑行人数。你需要用你的模型展示所有这些数据。

### 哑变量
将季节、天气、月份、假期、周天数转换成哑变量，因为这些变量的数字并没有实际的数量意义。

```python
dummy_fields = ['season', 'weathersit', 'mnth', 'hr', 'weekday']
for each in dummy_fields:
    dummies = pd.get_dummies(rides[each], prefix=each, drop_first=False)
    rides = pd.concat([rides, dummies], axis=1)

fields_to_drop = ['instant', 'dteday', 'season', 'weathersit', 
                  'weekday', 'atemp', 'mnth', 'workingday', 'hr']
data = rides.drop(fields_to_drop, axis=1)
data.head()
```

### 对一些变量进行标准化
为了使训练网络更容易，对一些连续变量进行标准化处理，即对这些变量的均值为0，标准差为1。
将变换系数存下来，在预测的时候使用。

```python
quant_features = ['casual', 'registered', 'cnt', 'temp', 'hum', 'windspeed']
# Store scalings in a dictionary so we can convert back later
scaled_features = {}
for each in quant_features:
    mean, std = data[each].mean(), data[each].std()
    scaled_features[each] = [mean, std]
    data.loc[:, each] = (data[each] - mean)/std
```

### 将数据分为训练集、验证集和测试集

我们将大约最后 21 天的数据保存为测试数据集，这些数据集会在训练完网络后使用。我们将使用该数据集进行预测，并与实际的骑行人数进行对比。

```python
# Save data for approximately the last 21 days 
test_data = data[-21*24:]

# Now remove the test data from the data set 
data = data[:-21*24]

# Separate the data into features and targets
target_fields = ['cnt', 'casual', 'registered']
features, targets = data.drop(target_fields, axis=1), data[target_fields]
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields]
```

我们将数据拆分为两个数据集，一个用作训练，一个在网络训练完后用来验证网络。因为数据是有时间序列特性的，所以我们用历史数据进行训练，然后尝试预测未来数据（验证数据集）。

```python
# Hold out the last 60 days or so of the remaining data as a validation set
train_features, train_targets = features[:-60*24], targets[:-60*24]
val_features, val_targets = features[-60*24:], targets[-60*24:]
```

## 构建网络的接口函数
  

```python
import tensorflow as tf

def get_features_input(n_features):
    """
    返回一个用于批量输入的tensor
    n_features：特征的数量
    return    ：占位符
    """
    return tf.placeholder(tf.float32, [None, n_features], name="inputs")

def get_targets_input():
    """
    返回用于targets的tensor
    return：占位符
    """
    return tf.placeholder(tf.float32, None, name="targets")

def get_learning_rate_input():
    """
    返回学习率的tensor
    learning_rate：学习率
    return       ：学习率的变量
    """
    return tf.placeholder(tf.float32, name="lr")

def add_layers(inputs, in_size, out_size, act_fun = None):
    """
    增加一个网络层
    input   ：输入
    in_size ：输入的节点数
    out_size：输出的节点数
    act_fun ：激活函数
    """
    #先计算线性输出
    weights = tf.Variable(tf.truncated_normal([in_size, out_size], stddev = 0.1))
    bias = tf.Variable(tf.zeros(out_size))
    wx_b = tf.matmul(inputs, weights) + bias
    
    #根据激活计算最终输出
    if act_fun == None:
        output = wx_b
    else:
        output = act_fun(wx_b)
    
    return output

def build_network(inputs, hidden_size, out_size, act_fun=tf.nn.sigmoid):
    """
    构建网络
    in_size     ：输入层节点数
    hidden_size ：隐藏层的技点数
    out_size    ：输出层的节点数
    lr          ：学习率
    act_fun     ：激活函数
    reutn       ：输出值、损失值、优化器、评价指标（RMSLE）
    """
    in_size = inputs.shape.as_list()[1]
    
    #增加层数，并计算输出
    #隐藏层使用激活函数，输出层不使用激活函数
    hidden_outputs = add_layers(inputs, in_size, hidden_size, act_fun)
    outputs = add_layers(hidden_outputs, hidden_size, out_size)
    
    return outputs
```

## 设置参数

```python
### 在这里设置参数###
iterations = 1
learning_rate = 0.0001
hidden_nodes = 30
output_nodes = 1
act_fun = tf.nn.sigmoid
n_features = train_features.shape[1]
```

## 创建Graph
该网络有两个层级，一个隐藏层和一个输出层。隐藏层级将使用 S 型函数作为激活函数。输出层只有一个节点，用于递归，节点的输出和节点的输入相同

```python
#定义输入
x = get_features_input(n_features)
y = get_targets_input()
lr = get_learning_rate_input()

#定义输出
prediction = build_network(x, hidden_nodes, output_nodes, act_fun)

#定义loss
loss = tf.reduce_mean(tf.square(prediction - y))

#定义优化方法
train_step = tf.train.GradientDescentOptimizer(lr).minimize(loss) 

#定义评价指标
evaluation = tf.sqrt(tf.reduce_sum(tf.square((tf.log(prediction-1) - tf.log(y-1)))))
```

## 训练网络

```python
import sys

losses = {'train':[], 'validation':[]}

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for ii in range(iterations):
        # 随机选择一个batch
        batch = np.random.choice(train_features.index, size=128)
        train_x, train_y = train_features.ix[batch].values, train_targets.ix[batch]['cnt']

        # 计算训练误差和验证误差
        train_loss,_ = sess.run([loss, train_step], feed_dict = {x:train_x, y:train_y, lr:learning_rate})
        valid_loss = sess.run(loss, feed_dict = {x:val_features, y:val_targets['cnt']})
        
        # 打印误差，并记录
        sys.stdout.write("\rProgress: {:2.1f}".format(100 * ii/float(iterations)) \
                         + "% ... Training loss: " + str(train_loss)[:5] \
                         + " ... Validation loss: " + str(valid_loss)[:5])
        sys.stdout.flush()

        losses['train'].append(train_loss)
        losses['validation'].append(valid_loss)
```

```python
plt.plot(losses['train'], label='Training loss')
plt.plot(losses['validation'], label='Validation loss')
plt.legend()
_ = plt.ylim()
```
